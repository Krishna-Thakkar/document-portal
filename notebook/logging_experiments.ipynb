```python
import logging
import os
from datetime import datetime
```

```python
logs_dir = os.path.join(os.getcwd(), "logs")
os.makedirs(logs_dir, exist_ok=True)
```

```python
LOG_FILE = f"{datetime.now().strftime('%m_%d_%Y_%H_%M_%S')}.log"
LOG_FILE
```

```python
LOG_FILE_PATH = os.path.join(logs_dir, LOG_FILE)
```

format = "[ %(asctime)s ] %(levelname)s %(name)s (line:%(lineno)d) - %(message)s"
format specifies the format of our log line.
1. asctime: date and time
2. levelname: INFO, WARNING, ERROR etc. (see above picture)
3. name: name of our project (DocumentPortal)
4. lineno: line number
5. message: message of our log line

example:
[ 2025-07-26 14:50:51,838 ] INFO DocumentPortal (line:1) - This is an info message

```python
# only logging INFO and above
logging.basicConfig(
    filename=LOG_FILE_PATH,
    format="[ %(asctime)s ] %(levelname)s %(name)s (line:%(lineno)d) - %(message)s",
    level=logging.INFO
)
```

```python
logger=logging.getLogger("DocumentPortal")
```

```python
logger.info("This is an info message")
```
